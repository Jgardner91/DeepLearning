## Initial Results 

Below is a training and validation accuracy plot over 70 epochs. These results came from a CNN model with transfer learning. Feature extraction was used, using the VGG16 pretrained network. This was then used as my convulational base in my model. 

The input, which was 1000 images each of Siberian and Scottish Fold cats were broken down into a 500/250/250 train/val/split. The initial result was an Accuracy Score on validation of 88.2%
